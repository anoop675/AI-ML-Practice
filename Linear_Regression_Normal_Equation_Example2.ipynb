```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import add_dummy_feature

np.random.seed(42) #ensures that the same sequence of "random" numbers will be produced every time the code is run with the same seed value
m = 100 #number of instances
X = np.random.rand(m, 1) * 2 #creates a 100x1 NumPy array (a column vector) filled with random floating-point numbers, each between 0 and 2 (exclusive of 2)
random_noise = np.random.rand(m, 1) #to simulate realistic data.
y = 4 + 3 * X + random_noise # here y is the target values, and the most optimal weight (slope) is 3, and most optimal bias (intercept) is 4

plt.figure(figsize=(10, 6))
plt.scatter(X, y)
plt.grid(True)
plt.xlabel("x (features)")
plt.ylabel("y (targets)")
plt.title("Linear data")
plt.show()

X_b = add_dummy_feature(X) #adds a new column of ones as the first column in the column vector 'X', forming a matrix
# because without the dummy feature (column vector of 1s) then we only consider optimizing the weights (since, the features are dot product) and not the bias
best_params = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y #Normal Equation returns the learned parameters that minimizes the MSE
print(f"Optimized weight: {best_params[1][0]} and Optimized bias: {best_params[0][0]}")
```
